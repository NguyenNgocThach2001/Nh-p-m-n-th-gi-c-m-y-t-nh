```python
import numpy as np
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("image/img.jpg", cv2.IMREAD_GRAYSCALE)
print(img.max())
plt.imshow(img, cmap='gray')
plt.show()
```

```python
from unicodedata import name


def derivative(X, direction):
    if(direction == 0):
        for idx in range(X.shape[0] - 1):
            X[idx,0:X.shape[1]] = X[idx, 0:X.shape[1]] - X[idx + 1, 0:X.shape[1]]
    else:
        for idx in range(X.shape[1] - 1):
            X[0:X.shape[0], idx] = X[0:X.shape[0], idx] - X[0:X.shape[0], idx + 1]
    return X
    
imgX = derivative(img.copy(), 0)
imgY = derivative(img.copy(), 1)

print("imgX", imgX.max())
print("imgY", imgY.max())
print("imgX", imgX.min())
print("imgY", imgY.min())


plt.imshow(imgX, cmap='gray')
plt.title("X")
plt.figure()

plt.title("Y")
plt.imshow(imgY, cmap='gray')
plt.show()

```

```python
def gradient_norm(X, Y):
    D = np.sqrt(X[:]*X[:]  + Y[:]*Y[:]) 
    return D

result = gradient_norm(imgX, imgY)
print(result)
plt.imshow(result, cmap='gray')
plt.show()
plt.figure()
plt.show()
```
